### Set style for graph

```python
# Setting style for bar graphs
import matplotlib.pyplot as plt
%matplotlib inline

# Set font
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'STIXGeneral'
fig, ax = plt.subplots(figsize=(8.5,11))

# Decorative Lines
ax.axvline(x=.5, ymin=0, ymax=1, color='#007ACC', alpha=0.0, linewidth=50)
plt.axvline(x=.99, color='#000000', alpha=0.5, linewidth=300)
plt.axhline(y=.88, xmin=0, xmax=1, color='#ffffff', linewidth=3)

# Set background color
ax.set_facecolor('white')

# Remove axes
plt.axis('off')

# Add Header Text
plt.annotate(Header, (.02,.98), weight='regular', fontsize=8, alpha=.75)
plt.annotate(Name, (.02,.94), weight='bold', fontsize=20)
plt.annotate(Title, (.02,.91), weight='regular', fontsize=14)

# Add Contact Info Text
plt.annotate(Contact, (.7,.906), weight='regular', fontsize=8, color='#ffffff')

# Add Summary Text
plt.annotate(SummaryHeader, (.02,.86), weight='bold', fontsize=10, color='#58C1B2')
plt.annotate(SummaryText, (.02,.76), weight='regular', fontsize=8)

# Add Professional Experience Text
plt.annotate(ExperienceHeader, (.02,.72), weight='bold', fontsize=10, color='#58C1B2')

# Job 1
plt.annotate(ExperienceOneTitle, (.02,.69), weight='bold', fontsize=10)
plt.annotate(ExperienceOneDate, (.02,.675), weight='regular', fontsize=8, alpha=.6)
plt.annotate(ExperienceOneDesc, (.04,.625), weight='regular', fontsize=9)

# Job 2
plt.annotate(ExperienceTwoTitle, (.02,.595), weight='bold', fontsize=10)
plt.annotate(ExperienceTwoDate, (.02,.58), weight='regular', fontsize=8, alpha=.6)
plt.annotate(ExperienceTwoDesc, (.04,.53), weight='regular', fontsize=9)

# Job 3
plt.annotate(ExperienceThreeTitle, (.02,.5), weight='bold', fontsize=10)
plt.annotate(ExperienceThreeDate, (.02,.485), weight='regular', fontsize=8, alpha=.6)
plt.annotate(ExperienceThreeDesc, (.04,.435), weight='regular', fontsize=9)

# Add Education Text
plt.annotate(EducationHeader, (.02,.395), weight='bold', fontsize=10, color='#58C1B2')

# Degree One
plt.annotate(DegreeOneTitle, (.02,.365), weight='bold', fontsize=10)
plt.annotate(CollegeOne, (.04,.35), weight='regular', fontsize=9)
plt.annotate(DegreeOneTime, (.04,.335), weight='regular', fontsize=9, alpha=.6)

# Degree Two
plt.annotate(DegreeTwoTitle, (.02,.305), weight='bold', fontsize=10)
plt.annotate(CollegeTwo, (.04,.29), weight='regular', fontsize=9)
plt.annotate(DegreeTwoTime, (.04,.275), weight='regular', fontsize=9, alpha=.6)

# Degree Three
plt.annotate(DegreeThreeTitle, (.02,.245), weight='bold', fontsize=10)
plt.annotate(CollegeThree, (.04,.23), weight='regular', fontsize=9)
plt.annotate(DegreeThreeTime, (.04,.215), weight='regular', fontsize=9, alpha=.6)

# Degree Four
plt.annotate(DegreeFourTitle, (.02,.185), weight='bold', fontsize=10)
plt.annotate(CollegeFour, (.04,.17), weight='regular', fontsize=9)
plt.annotate(DegreeFourTime, (.04,.155), weight='regular', fontsize=9, alpha=.6)

# Skills 
plt.annotate(SkillsHeader, (.7,.8), weight='bold', fontsize=10, color='#ffffff')
plt.annotate(SkillsDesc, (.7,.51), weight='regular', fontsize=10, color='#ffffff')
#plt.annotate(ExtrasTitle, (.7,.43), weight='bold', fontsize=10, color='#ffffff')
#plt.annotate(ExtrasDesc, (.7,.345), weight='regular', fontsize=10, color='#ffffff')
#plt.annotate(CodeTitle, (.7,.2), weight='bold', fontsize=10, color='#ffffff')

from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox
import matplotlib.image as mpimg
plt.savefig('resumeexample1.png', dpi=300, bbox_inches='tight')
```
